## Assignment 1: Analyze Relationships  
Determining which factors most strongly correlate with a track's success.

```python
df1_heatmap = df1[["track_popularity", "artist_popularity", "artist_followers", "track_duration_min", "album_total_tracks"]]
```

```python
correlation_matrix = df1_heatmap.corr(method="spearman")

sns.heatmap(correlation_matrix, annot=True, fmt=".3f", linewidth=.5, cmap="Greens")

plt.savefig("track_popularity correlation.png")
```

Based on the heatmap we can see there's correlation between artist_popularity and artist_followers  
that is associated with a track gaining in popularity.

```python
sns.regplot(data=df1, x="artist_popularity", y="track_popularity", color="green", line_kws=dict(color="black"));

plt.savefig("track vs artist popularity.png")
```

## Assignment 2: The Evolution of Music  
Visualizing how music production trends have changed over the decades.

```python
df1["album_release_date"] = df1["album_release_date"].dt.year
```

```python
df1_2009_2025 = (df1["album_release_date"] >=2009) & (df1["album_release_date"] <=2025)
```

```python
avg_track_duration = df1[df1_2009_2025].groupby("album_release_date")["track_duration_min"].mean().reset_index()
```

```python
volume_of_releases = df1[df1_2009_2025].groupby("album_release_date")["album_release_date"].count()
```

```python
figure, ax1 = plt.subplots(figsize=(6,4))
ax2 = ax1.twinx() #Matplotlib function that creates a second y-axis that shares the same x-axis as the main plot.

### Avg. Track Duration
sns.set_style("whitegrid")
sns.lineplot(data=avg_track_duration, x="album_release_date", y="track_duration_min", marker="o",
             color = "green", ax=ax1, label="Avg. Track Duration")

if ax1.legend_: #Removes seaborn auto legend
    ax1.legend_.remove()
ax1.set_xlabel("Album Release")
ax1.set_ylabel("Avg. Track Duration(minutes)")


sns.lineplot(data=volume_of_releases, marker="o",
             color = "black", ax=ax2, label="Volume of Releases")

plt.title("Evolution of Music (2009-2025)")

if ax2.legend_: #Removes seaborn auto legend
    ax2.legend_.remove()
ax2.set_xlabel("Album Release")
ax2.set_ylabel("Volume of Releases")

# Manually creating the legend
lines = ax1.get_lines() + ax2.get_lines()
labels = [line.get_label() for line in lines]
ax1.legend(lines, labels, loc="center left")

plt.savefig("Evolution of Music.png")

plt.tight_layout()
plt.show()
```

## Assignment 3: Genres  
Identifying the most prevalent genres.

```python
df1.head()
```

```python
df1_genres = df1.dropna(subset="artist_genres", axis=0)
```

```python
# .str.split() breaks a string into a list of substrings based on a specified delimiter
df1_genres["artist_genres"] = df1_genres["artist_genres"].str.split(", ")

# .explode() used with pandas library, transforms each element of a list-like column into a separate row
df1_genres = df1_genres.explode("artist_genres")
```

```python
df1_genre_count = df1_genres["artist_genres"].value_counts().head(15)
```

```python
plt.figure(figsize=(15,6))

sns.set_style("whitegrid")
sns.barplot(data=df1_genre_count, color = "green")

plt.title("Top 15 Genres")
plt.xlabel("Genres")
plt.ylabel("Count")

plt.savefig("Top 15 Genres.png")

plt.show();
```
